```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import seaborn as sns
#한글 설정
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False) # y축

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/1)재택활동지수_2000921 (1).csv')
df
```

```python
df['연월'] = df['dt'].apply(lambda x:str(x)[2:4]+'.'+str(x)[4:6])
df
```

```python
df1 = df.loc[:,['연월', 'sex_cd', 'home_sido_nm', 'h0d1h1_dur_r', 'h1d1h1_dur_r']]
df1.columns=['연월', '성별', '거주지역', '평일주간재택지수', '휴일주간재택지수']
df1
```

```python
df1.groupby('연월').count()
```

```python
df1['연도'] = df1['연월'].apply(lambda x:x[:2])
df1
```

```python
set(df1['거주지역'])
```

```python
df2 = df1.groupby(['거주지역', '연도']).mean()
df2.reset_index(inplace=True)
df2.drop(columns=['성별'], inplace=True)
df2
```

```python
plt.figure(figsize=(20,5))
sns.pointplot(x='거주지역', y='평일주간재택지수', hue='연도', data=df2)
plt.title('평일 주간 재택지수')
plt.savefig('data2/지역별_평일주간재택지수.jpg')
```

```python
plt.figure(figsize=(20,5))
sns.pointplot(x='거주지역', y='휴일주간재택지수', hue='연도', data=df2)
plt.title('휴일 주간 재택지수')
plt.savefig('data2/지역별_휴일주간재택지수.jpg')
```

```python
df1.groupby(['거주지역', '연월']).mean()
```

```python
df1_mean = df1.groupby(['연월', '성별']).mean()
df1_mean.reset_index(inplace=True)
```

```python
fig, ax1 = plt.subplots(figsize=(15,5))
ax2 = ax1.twinx()
ax1 = sns.barplot(x='연월', y='평일주간재택지수', data=df1_mean, ci=None, ax=ax1)
ax2 = sns.pointplot(x='연월', y='평일주간재택지수', hue='성별', data=df1_mean, ax=ax2)
ax2.legend(loc='upper left', labels=['남성'])
plt.title('평일주간재택지수')
```

```python
fig, ax1 = plt.subplots(figsize=(15,5))
ax2 = ax1.twinx()
ax1 = sns.barplot(x='연월', y='휴일주간재택지수', data=df1_mean, ci=None, ax=ax1)
ax2 = sns.pointplot(x='연월', y='휴일주간재택지수', hue='성별', data=df1_mean, ax=ax2)
ax2.legend(loc='upper left', labels=['남성'])
plt.title('휴일주간재택지수')
```
