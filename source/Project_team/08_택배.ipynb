```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import seaborn as sns
#한글 설정
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False) # y축

import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/생활물류통계_매출액_이용횟수.xlsx')
df
```

```python
fig, ax1 = plt.subplots(figsize=(15,5))
ax2 = ax1.twinx()
ax1 = sns.barplot(x='연도', y='매출액(억원)', data=df, ax=ax1, palette='husl')
ax2 = sns.lineplot(x='연도', y='물동량(만박스)', data=df, ax=ax2)
for i in range(df.shape[0]):
    ax1.text(x=i, y=df['매출액(억원)'][i], s=df['매출액(억원)'][i], horizontalalignment='center')
plt.title('국내 택배시장 총매출액 및 물동량')
plt.savefig('data2/연도별_국내택배시장_총매출액_물동량.jpg')
```
